```python
import pandas as pd
import sqlite3

# Create or connect to a SQLite database
conn = sqlite3.connect("walmart.db")
```

```python
sales = pd.read_csv("train.csv")
features = pd.read_csv("features.csv")
stores = pd.read_csv("stores.csv")
```

```python
sales.to_sql("Sales_Transactions", conn, if_exists="replace", index=False)
features.to_sql("Features", conn, if_exists="replace", index=False)
stores.to_sql("Stores", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT Store, Date, SUM(Weekly_Sales) AS Total_Sales
FROM Sales_Transactions
GROUP BY Store, Date
ORDER BY Date, Store
LIMIT 10;
"""
df_result = pd.read_sql_query(query, conn)
df_result
```

```python
query = """
SELECT IsHoliday, AVG(Weekly_Sales) AS Avg_Sales
FROM Sales_Transactions
GROUP BY IsHoliday;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT Store, Date, Weekly_Sales,
       AVG(Weekly_Sales) OVER (
           PARTITION BY Store
           ORDER BY Date
           ROWS BETWEEN 3 PRECEDING AND CURRENT ROW
       ) AS Moving_Avg_Sales
FROM Sales_Transactions
LIMIT 20;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    s.Store,
    s.Dept,
    s.Date,
    s.Weekly_Sales,
    s.IsHoliday,
    f.Temperature,
    f.Fuel_Price,
    f.CPI,
    f.Unemployment,
    st.Type AS Store_Type,
    st.Size AS Store_Size
FROM Sales_Transactions s
JOIN Features f
  ON s.Store = f.Store AND s.Date = f.Date
JOIN Stores st
  ON s.Store = st.Store
ORDER BY s.Store, s.Date;
"""
df_joined = pd.read_sql_query(query, conn)
print("Joined dataset shape:", df_joined.shape)
```

```python
print("Joined dataset shape:", df_joined.shape)
df_joined.head()
```

```python
df_joined["Lag_1"] = df_joined.groupby(["Store","Dept"])["Weekly_Sales"].shift(1)
df_joined["Lag_2"] = df_joined.groupby(["Store","Dept"])["Weekly_Sales"].shift(2)
df_joined["Lag_3"] = df_joined.groupby(["Store","Dept"])["Weekly_Sales"].shift(3)
df_model = df_joined.dropna()
print("After lagging:", df_model.shape)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error

# Features
X = df_model[["Lag_1", "Lag_2", "Lag_3", "Temperature", "Fuel_Price", "CPI", "Unemployment", "Store_Size"]]
y = df_model["Weekly_Sales"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Model
model = XGBRegressor(n_estimators=200, learning_rate=0.1, max_depth=6)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print("MAPE:", mean_absolute_percentage_error(y_test, y_pred))
```

```python
print("Min Weekly Sales:", y_test.min())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print("RMSE:", rmse)
print("MAE:", mae)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(y_test.values[:100], label="Actual", color="blue")
plt.plot(y_pred[:100], label="Predicted", color="red")
plt.legend()
plt.title("Weekly Sales Forecast vs Actual")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Predicted vs Actual Scatter Plot")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')  # perfect line
plt.show()
```

```python
residuals = y_test.values - y_pred
plt.figure(figsize=(12,6))
plt.plot(residuals[:100], label="Residuals", color="purple")
plt.axhline(0, color="black", linestyle="--")
plt.legend()
plt.title("Residuals Over Time")
plt.show()
```
